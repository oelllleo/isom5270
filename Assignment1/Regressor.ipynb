```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Labeled_data.csv')
df.info()
```

```python
df.head(10)
```

```python
# dummary: state, zip, rollout, collscr
df = pd.get_dummies(df,  columns=['state', 'zip', 'rollout','collscr'], drop_first=True)
df.head(10)
```

```python
X = df.drop(columns = ['acctid', 'totalpay'])
y = df.totalpay.values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error, r2_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

rf = RandomForestRegressor(max_depth= 6, min_samples_split= 15, min_samples_leaf= 5, random_state= 1337)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
n = len(y_test)  # Number of observations
p = X_test.shape[1]  # Number of predictors

# Calculate Adjusted R-squared
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Print results
print("Mean Absolute Percentage Error:", mape)
print("R-squared:", r2)
print("Adjusted R-squared:", adjusted_r2)
```

```python
X_test = X_test.reset_index().drop(columns = 'index')
buyornot_pred = []
buyornot_test = []
for i in range(len(y_test)):
    if (y_pred[i] - X_test['cobal'][i]*0.25) <0:
        buy_pred = 0
    else:
        buy_pred = 1
    buyornot_pred.append(buy_pred)
    
    if (y_test[i] - X_test['cobal'][i]*0.25) <0:
        buy_test = 0
    else:
        buy_test = 1
    buyornot_test.append(buy_test)
    i = i+1
```

#### The confusion matrix: 

```python
tn, fp, fn, tp = confusion_matrix(buyornot_test, buyornot_pred).ravel()
(tn, fp, fn, tp)
```

#### Thus, FP = 269/714 = 0.3767 ~ 38%, FN = 56/714 = 0.07843 ~ 8%

```python
### using y_pred & buyornot_pred to calculate expected value on test group

X_test['decision'] = buyornot_pred
X_test['profit'] = list(map(lambda x:max(y_test[x]-0.25*(X_test['cobal'][x]), -0.25*(X_test['cobal'][x])), range(len(y_test))))

X_test['expected_profit'] = X_test['decision']*X_test['profit']
```

```python
X_test[X_test['expected_profit']<0]
```

```python
X_test.loc[:,['decision','expected_profit']].describe()
```

```python
mean_profit = np.mean(X_test['expected_profit'])

# Step 3: Calculate the standard deviation
std_dev = np.std(X_test['expected_profit'], ddof=1)  # Sample standard deviation

# Step 4: Calculate the sample size
n = len(X_test['expected_profit'])

# Step 5: Calculate the margin of error using z-distribution
z_critical = 1.96  # Critical value for 95% CI
margin_of_error = z_critical * (std_dev / np.sqrt(n))  # Standard error is std_dev / sqrt(n)

# Step 6: Calculate the confidence interval
confidence_interval = (mean_profit - margin_of_error, mean_profit + margin_of_error)

# Output results
print(f"Mean Profit: {mean_profit:.2f}")
print(f"95% Confidence Interval for Expected Profit (Z-distribution): {confidence_interval}")
```
